```python
# Transformasi
# Membersihkan data dari duplikat
duplicate_rows = data_survey[data_survey.duplicated()]
data_survey = data_survey.drop_duplicates()
duplicate_rows
```

```python
# Membersihkan data dari baris yang memiliki nilai kosong
data_survey = data_survey.dropna()
```
