```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
electronic = pd.read_csv("electronic_shop.csv")
electronic
```

## Membuat sebuah line plot untuk menggambarkan jumlah nilai dari transaksi per hari.

```python
# Mengelompokkan data berdasarkan tanggal dan menjumlahkan nilai transaksi
daily_transactions = electronic.groupby('transaction_date')['transaction_amount'].sum().reset_index()
```

```python
# Membuat line plot
plt.figure(figsize=(10, 6))
plt.plot(daily_transactions['transaction_date'], daily_transactions['transaction_amount'], marker='o')
plt.title('Jumlah Nilai Transaksi per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Nilai Transaksi')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

## Membuat sebuah bar plot untuk menggambarkan jumlah customer per hari.

```python
# Mengelompokkan data berdasarkan tanggal dan menghitung jumlah unique customer
daily_customers = electronic.groupby('transaction_date')['transaction_id'].nunique().reset_index()
```

```python
# Membuat bar plot
plt.figure(figsize=(10, 6))
plt.bar(daily_customers['transaction_date'], daily_customers['transaction_id'])
plt.title('Jumlah Customer per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Customer')
plt.xticks(rotation=45)
plt.show()
```

## Membuat sebuah scatter plot untuk menggambarkan jumlah nilai dari transaksi per hari.

```python
# Mengelompokkan data berdasarkan tanggal dan menjumlahkan nilai transaksi (sama seperti line plot)
daily_transactions = electronic.groupby('transaction_date')['transaction_amount'].sum().reset_index()
```

```python
# Membuat scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(daily_transactions['transaction_date'], daily_transactions['transaction_amount'], color='red')
plt.title('Scatter Plot Jumlah Nilai Transaksi per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Nilai Transaksi')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
