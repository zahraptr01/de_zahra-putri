```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bakery = pd.read_csv("bakery.csv")
bakery
```

## Jumlah nilai transaksi per hari.

```python
# Mengelompokkan data berdasarkan tanggal dan menjumlahkan nilai transaksi
daily_transactions = bakery.groupby('transaction_date')['transaction_amount'].sum().reset_index()
```

```python
# Membuat line plot
plt.figure(figsize=(10, 6))
plt.plot(daily_transactions['transaction_date'], daily_transactions['transaction_amount'], marker='o')
plt.title('Jumlah Nilai Transaksi per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Nilai Transaksi')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

## Jenis roti yang paling banyak dipesan

```python
# Menghitung jumlah pesanan untuk setiap jenis roti
bread_counts = bakery['item_name'].value_counts()
```

```python
# Menampilkan jenis roti yang paling banyak dipesan
most_ordered_bread = bread_counts.idxmax()
most_ordered_bread_count = bread_counts.max()

print(f"Jenis roti yang paling banyak dipesan adalah: {most_ordered_bread} dengan jumlah pesanan: {most_ordered_bread_count}")
```

## Rata-rata kepuasan pelanggan (customer_rating) per hari

```python
# Mengelompokkan data berdasarkan tanggal dan menghitung rata-rata customer rating
daily_avg_rating = bakery.groupby('transaction_date')['customer_rating'].mean().reset_index()
```

```python
# Membuat bar plot untuk rata-rata kepuasan pelanggan per hari
plt.figure(figsize=(10, 6))
plt.bar(daily_avg_rating['transaction_date'], daily_avg_rating['customer_rating'])
plt.title('Rata-rata Kepuasan Pelanggan per Hari')
plt.xlabel('Tanggal')
plt.ylabel('Rata-rata Rating Kepuasan')
plt.xticks(rotation=45)
plt.show()
```
