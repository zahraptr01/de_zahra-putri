```python
import pandas as pd
import csv

# Load data csv
df = pd.read_csv('logs.csv')

def process_data(df):
    # Cek missing values
    print("Missing values:")
    print(df.isnull().sum())
    
    # Cek duplikat data
    print("Duplicate rows:")
    print(df[df.duplicated()])
    
    # Hapus duplikat data
    df = df.drop_duplicates()
    
    return df

# Buat data frame
open_product = df[df['textPayload'].str.contains('OPEN-CATEGORY')]
search_product = df[df['textPayload'].str.contains('SEARCH-PRODUCT')]
open_category = df[df['textPayload'].str.contains('OPEN-CATEGORY')]

```

```python
# Proses data
open_product = process_data(open_product)
search_product = process_data(search_product)
open_category = process_data(open_category)
```

```python
# Simpan data ke file CSV
open_product.to_csv('transformed_data/open_product.csv', index=False)
search_product.to_csv('transformed_data/search_product.csv', index=False)
open_category.to_csv('transformed_data/open_category.csv', index=False)
```

```python
with open('transformed_data/open_product.csv', 'r') as file:
    reader = csv.DictReader(file)
    data = [row for row in reader]

transformed_data = []
for row in data:
    timestamp = row['timestamp']
    log_start = '[LOG_DE_START]'
    user_id = row['textPayload'].split()[1]
    action = row['textPayload'].split()[2]
    product_id = ' '.join(row['textPayload'].split()[3:-1])
    log_end = '[LOG_DE_END]'

    transformed_row = {
        'timestamp': timestamp,
        'log_start': log_start,
        'user_id': user_id,
        'action': action,
        'product_id': product_id,
        'log_end': log_end
    }
    transformed_data.append(transformed_row)

open_product = pd.DataFrame(transformed_data)
open_product.to_csv('transformed_data/open_product2.csv', index=False)
open_product = open_product.drop(['log_start', 'action', 'log_end'], axis=1)
open_product.to_csv('transformed_data/open_product3.csv', index=False)
df_op3 = pd.read_csv('transformed_data/open_product3.csv')
df_op3
```
