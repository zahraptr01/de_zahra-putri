```python
import pandas as pd
```

```python
tx_data = pd.read_csv('tx_data.csv')
tx_data
```

```python
# Mengubah format nomor telepon dengan awalan +62 di depan. 
# Contohnya seperti ini: +6281234567891
tx_data['phone_number'] = '+62' + tx_data['phone_number'].astype(str)
tx_data.head()
```

```python
# Mengubah nilai transaksi (transaction_amount) dengan format seperti berikut: Rp 12.000
tx_data['transaction_amount'] = 'Rp ' + tx_data['transaction_amount'].astype(str).str.replace(',', '.')
tx_data
```

```python
# Pastikan data yang diambil adalah data transaksi dengan status sama dengan success.
tx_data = tx_data[tx_data['transaction_status'] == 'success']
tx_data
```

```python
tx_data.to_csv('C:/Users/ASUS/de_zahra-putri/19_Pemahaman tentang Data Quality dan Data Governance/Praktikum/data_cleaned.csv')
df = pd.read_csv('data_cleaned.csv')
df
```

```python
# Hasil akhir data cleaning disimpan di dalam DataFrame.
cleaned_df = pd.DataFrame({
    'transaction_id': df['transaction_id'],
    'customer_name': df['customer_name'],
    'transaction_amount': df['transaction_amount'],
    'phone_number': df['phone_number'],
    'transaction_type': df['transaction_type'],
    'transaction_status': df['transaction_status']
})
```

```python
# Simpan kembali hasil akhir data cleaning ke dalam file CSV
cleaned_df.to_csv('cleaned_tx_data.csv', index=False)
```
